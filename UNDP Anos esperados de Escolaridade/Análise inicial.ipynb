### Análise do dataset
O dataset trabalhado foi originado do Kaggle, entitulado Years of Schooling Worldwide, disponibilizado pelo usuários 'iamsouravbanerjee' em formato .csv(-https://www.kaggle.com/datasets/iamsouravbanerjee/years-of-schooling-worldwide)

Realizei o download do dataset através da API do Kaggle, e fiz uma análise inicial com Python, visando verificar as informações disponibilizadas e realizar uma análise prévia.

```python
# bibliotecas utlizadas 
import pandas as pd
import numpy as np

import kaggle                                           
from kaggle.api.kaggle_api_extended import KaggleApi 
```

```python
#realizando o download do arquivo.csv (resposta False pois já havia realizado o download anteriormente utilizando o mesmo processo)
api = KaggleApi()
api.authenticate()

api.dataset_download_file('iamsouravbanerjee/years-of-schooling-worldwide', 'schooling.csv')
```

```python
df = pd.read_csv('schooling.csv')
df.head()
```

```python
#Verificação do conteúdo do dataset
df.info()
```

#### Análise de dados nulos
Podemos verificar que existem dados faltantes em diversas colunas:

- Human Development Groups;
- UNDP Developing Regions;
- HDI Rank (2021);
- Expected Years of Schooling(1990 - 2021)

Abaixo analisei cada caso separadamente.

#### Human Development Groups

```python
df.loc[df['Human Development Groups'].isna() ]
```

```python
df['Human Development Groups'].unique()
```

Em pesquisa ao site do "Human Development Report", constatei que os 4 países que não apresentam dados de 'Human Development Groups' também não estão descritos no site (https://hdr.undp.org/data-center/country-insights#/ranks). Para evitar erros de classificação, deixarei o valor NaN.

### HDI Rank (2021)

```python
df.loc[df['HDI Rank (2021)'].isna() ]
```

Verifiquei que os mesmos 4 países não apresentam rank de Índice de Desenvolvimento Humano, pois não estão descritos no Human Development Reports (HDR) do Programa de Desenvolvimento das Nações Unidas (UNDP). Assim, não tenho como apresentar esta informação e portanto deixarei como NaN.

### UNDP Developing Regions

```python
df.loc[df['UNDP Developing Regions'].isna() ]
```

Analisei a coluna 'UNDP Development Region' e pude constatar que ela é aplicada à países em desenvolvimento. Assim, dos países presentes no dataset, aqueles considerados 'desenvolvidos' não estão alocados em nenhuma região.

### Expected Years of Schooling(1990 - 2021)

```python
df.loc[df['Expected Years of Schooling (1990)'].isna() ]
```

Verifiquei os dados faltantes referentes aos anos de 1990 até 2021. A ausencia de informação pode servir de comparação com outros países, portanto manterei os dados no datase.
